# ONNX to Keras

```python
import tensorflow as tf
import onnx
from onnx2keras import onnx_to_keras

# Load ONNX model
onnx_model = onnx.load('models/state_dicts/resnet18.onnx')

# Call the converter (input - is the main model input name, can be different for your model)
k_model = onnx_to_keras(onnx_model, ['input'])
```

```python
# tf.saved_model.save(k_model, "models/state_dicts/resnet18h5")
tf.keras.models.save_model(k_model, "models/state_dicts/resnet18h5")
```

```python
dnn = tf.keras.models.load_model("models/state_dicts/resnet18h5")
dnn.compile()
```
